```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Function to get image sizes
def get_image_sizes(dataset_dir):
    widths = []
    heights = []
    
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):  # Ensure it's a directory
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                try:
                    with Image.open(image_path) as img:
                        width, height = img.size
                        widths.append(width)
                        heights.append(height)
                except Exception as e:
                    print(f"Error processing {image_path}: {e}")
    
    return widths, heights

# Function to plot size distribution
def plot_size_distribution(widths, heights):
    plt.figure(figsize=(12, 6))
    
    # Plot histogram of widths
    plt.subplot(1, 2, 1)
    plt.hist(widths, bins=50, color='blue', alpha=0.7)
    plt.title('Distribution of Image Widths')
    plt.xlabel('Width (pixels)')
    plt.ylabel('Frequency')
    
    # Plot histogram of heights
    plt.subplot(1, 2, 2)
    plt.hist(heights, bins=50, color='green', alpha=0.7)
    plt.title('Distribution of Image Heights')
    plt.xlabel('Height (pixels)')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()

# Function to print summary statistics
def print_summary_statistics(widths, heights):
    print("Summary Statistics:")
    print(f"Min Width: {np.min(widths)}, Max Width: {np.max(widths)}")
    print(f"Min Height: {np.min(heights)}, Max Height: {np.max(heights)}")
    print(f"Mean Width: {np.mean(widths):.2f}, Mean Height: {np.mean(heights):.2f}")
    print(f"Median Width: {np.median(widths)}, Median Height: {np.median(heights)}")

# Main script
if __name__ == "__main__":
    dataset_dir = "/Users/laumartin4/code/Laumartin4/project_draft/data"  # Replace with your dataset path
    
    # Get image sizes
    widths, heights = get_image_sizes(dataset_dir)
    
    # Plot size distribution
    plot_size_distribution(widths, heights)
    
    # Print summary statistics
    print_summary_statistics(widths, heights)
```

```python
import os
import cv2
import numpy as np

def resize_with_padding(image, target_size=(224, 224)):
    """
    Resize an image while preserving the aspect ratio and adding padding.
    """
    h, w = image.shape[:2]
    target_w, target_h = target_size
    
    # Calculate scaling factor
    scale = min(target_w / w, target_h / h)
    
    # Resize while preserving aspect ratio
    resized_image = cv2.resize(image, None, fx=scale, fy=scale, interpolation=cv2.INTER_AREA)
    
    # Add padding
    pad_w = (target_w - resized_image.shape[1]) // 2
    pad_h = (target_h - resized_image.shape[0]) // 2
    padded_image = cv2.copyMakeBorder(
        resized_image, 
        pad_h, pad_h, pad_w, pad_w, 
        cv2.BORDER_CONSTANT, value=(0, 0, 0)  # Black padding
    )
    
    return padded_image

def resize_25_images_per_category(input_dir, output_dir, target_size=(224, 224), num_images=25):
    """
    Resize exactly 25 images from each category in the dataset.
    Ensure all resized images are of the target size (224x224).
    """
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for class_name in os.listdir(input_dir):
        class_path = os.path.join(input_dir, class_name)
        if os.path.isdir(class_path):  # Ensure it's a directory
            output_class_dir = os.path.join(output_dir, class_name)
            if not os.path.exists(output_class_dir):
                os.makedirs(output_class_dir)
            
            # Initialize counter for valid images
            valid_images_count = 0
            
            # Iterate through images in the category
            for image_name in os.listdir(class_path):
                if valid_images_count >= num_images:
                    break  # Stop after saving 25 valid images
                
                image_path = os.path.join(class_path, image_name)
                output_image_path = os.path.join(output_class_dir, image_name)
                
                # Load the image
                image = cv2.imread(image_path)
                if image is None:
                    print(f"Error loading {image_path}. Skipping...")
                    continue
                
                # Resize the image
                resized_image = resize_with_padding(image, target_size)
                
                # Check if the resized image has the correct dimensions
                if resized_image.shape[0] != target_size[0] or resized_image.shape[1] != target_size[1]:
                    print(f"Skipping {image_name}: Resized image is not {target_size}.")
                    continue
                
                # Save the resized image
                cv2.imwrite(output_image_path, resized_image)
                valid_images_count += 1
                print(f"Resized and saved: {output_image_path} ({valid_images_count}/{num_images})")
            
            # Print summary for the category
            print(f"Finished processing {class_name}. Saved {valid_images_count} valid images.")

# Example usage
input_dir = "/Users/laumartin4/code/Laumartin4/smearly/raw_data"
output_dir = "/Users/laumartin4/code/Laumartin4/smearly/resized_data_224_20"
resize_25_images_per_category(input_dir, output_dir, target_size=(224, 224))
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Function to get image sizes per class
def get_image_sizes_per_class(dataset_dir):
    class_sizes = {}
    
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):  # Ensure it's a directory
            widths = []
            heights = []
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                try:
                    with Image.open(image_path) as img:
                        width, height = img.size
                        widths.append(width)
                        heights.append(height)
                except Exception as e:
                    print(f"Error processing {image_path}: {e}")
            class_sizes[class_name] = (widths, heights)
    
    return class_sizes

# Function to calculate the maximum frequency across all classes
def get_max_frequency(class_sizes):
    max_freq_width = 0
    max_freq_height = 0
    
    for widths, heights in class_sizes.values():
        # Calculate histograms for widths and heights
        hist_width, _ = np.histogram(widths, bins=50)
        hist_height, _ = np.histogram(heights, bins=50)
        
        # Update max frequencies
        if np.max(hist_width) > max_freq_width:
            max_freq_width = np.max(hist_width)
        if np.max(hist_height) > max_freq_height:
            max_freq_height = np.max(hist_height)
    
    return max_freq_width, max_freq_height

# Function to plot size distribution per class with the same y scale
def plot_size_distribution_per_class(class_sizes, max_freq_width, max_freq_height):
    for class_name, (widths, heights) in class_sizes.items():
        plt.figure(figsize=(12, 6))
        
        # Plot histogram of widths
        plt.subplot(1, 2, 1)
        plt.hist(widths, bins=50, color='blue', alpha=0.7)
        plt.title(f'Distribution of Image Widths for {class_name}')
        plt.xlabel('Width (pixels)')
        plt.ylabel('Frequency')
        plt.ylim(0, max_freq_width)  # Set y-axis limit
        
        # Plot histogram of heights
        plt.subplot(1, 2, 2)
        plt.hist(heights, bins=50, color='green', alpha=0.7)
        plt.title(f'Distribution of Image Heights for {class_name}')
        plt.xlabel('Height (pixels)')
        plt.ylabel('Frequency')
        plt.ylim(0, max_freq_height)  # Set y-axis limit
        
        plt.tight_layout()
        plt.show()

# Function to print summary statistics per class
def print_summary_statistics_per_class(class_sizes):
    for class_name, (widths, heights) in class_sizes.items():
        print(f"Summary Statistics for {class_name}:")
        print(f"Min Width: {np.min(widths)}, Max Width: {np.max(widths)}")
        print(f"Min Height: {np.min(heights)}, Max Height: {np.max(heights)}")
        print(f"Mean Width: {np.mean(widths):.2f}, Mean Height: {np.mean(heights):.2f}")
        print(f"Median Width: {np.median(widths)}, Median Height: {np.median(heights)}")
        print()

# Main script
if __name__ == "__main__":
    dataset_dir = "/Users/laumartin4/code/Laumartin4/project_draft/data"  # Replace with your dataset path
    
    # Get image sizes per class
    class_sizes = get_image_sizes_per_class(dataset_dir)
    
    # Calculate the maximum frequency for widths and heights
    max_freq_width, max_freq_height = get_max_frequency(class_sizes)
    
    # Plot size distribution per class with the same y scale
    plot_size_distribution_per_class(class_sizes, max_freq_width, max_freq_height)
    
    # Print summary statistics per class
    print_summary_statistics_per_class(class_sizes)
```
